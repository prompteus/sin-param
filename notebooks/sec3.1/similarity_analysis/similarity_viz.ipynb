```python
def to_data(str_data):
    return pd.DataFrame.from_records([
        {
            'modelA': modelA, 
            'modelB': modelB, 
            'corr_num': float(corr_num), 
            'pval_num': float(pval_num),  
            'corr_bas': float(corr_bas), 
            'pval_bas': float(pval_bas), 
        }
        for modelA, modelB, corr_num, pval_num, corr_bas, pval_bas 
        in map(str.split, str_data.split('\n'))
    ] + [
        {
            'modelA': modelB, 
            'modelB': modelA, 
            'corr_num': float(corr_num), 
            'pval_num': float(pval_num),  
            'corr_bas': float(corr_bas), 
            'pval_bas': float(pval_bas),
        }
        for modelA, modelB, corr_num, pval_num, corr_bas, pval_bas 
        in map(str.split, str_data.split('\n'))
    ])

cosine_sim_df = to_data(cosine_sim)
euclidian_dist_df = to_data(euclidian_dist)
dot_product_df = to_data(dot_product)
```
